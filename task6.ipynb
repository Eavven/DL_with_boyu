机器学习中,由于样本特征取值的分布范围往往差异较大,在计算欧式距离时,范围大的特征会起到主导作用,因此要对特征进行归一化.

虽然神经网络可以通过参数的调整来适应不同特征的取值范围，但是会导致训练效率比较低.

原因一:避免为产生梯度爆炸或消失的人工干预,如tanh函数,其导数在区间[-2,2]是敏感的,当输入过大或过下,其梯度会趋向于0,难以训练.例如tanh(w1x1+w2x2+b),特征x1取值范围以千计,则w1必须初始化0.001这样的大小,否则梯度过小将难以训练.因此训练前先将特征归一化,例如范围[-1,1],这样无需区别对待w
原因二:，不同输入特征的取值范围差异比较大时，梯度下降法的效率也会受到影响.

机器学习中的协变量可以看作是输入，一般的机器学习算法都要求输入在训练集和测试集上的分布是相似的。而在深度神经网络中，由于该层的输入是上一层的输出，当使用随机梯度下降使得参数更新后，上一层的输出也会变化，该层的输入分布发生改变，那么参数需要重新学习，即协变量偏移。逐层归一化是为了解决内部斜变量偏移问题

批量归一化（Batch Normalization，BN）虽然归一化操作可以应用在输入𝒂(𝑙−1) 上，但其分布性质
不如𝒛(𝑙) 稳定. 因此，在实践中归一化操作一般应用在仿射变换之后，激活函数之
前.逐层归一化需要在中间层进行操作，
要求效率比较高，因此复杂度比较高的白化方法就不太合适. 为了提高归一化效
率，一般使用标准归一化

线性性质.
